## Importing Dependancies

```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
print(stopwords.words('English'))
```

## Importing Dataset

```python
news_data=pd.read_csv("WELFake_Dataset.csv")
```

## Data Analysis 

```python
news_data.head()
```

```python
news_data.shape
```

```python
news_data.isnull().sum()
```

```python
news_data=news_data.fillna('')
```

```python
news_data.isnull().sum()
```

```python
news_data.head()
```

## Data Preprocessing

```python
# Sample 17,000 rows with label 0 and 17,000 rows with label 1
rows_to_drop_0 = news_data[news_data['label'] == 0].sample(n=30000, random_state=42).index
rows_to_drop_1 = news_data[news_data['label'] == 1].sample(n=30000, random_state=42).index

# Drop these rows from the DataFrame
news_data = news_data.drop(index=rows_to_drop_0)
news_data = news_data.drop(index=rows_to_drop_1)

# Reset index if needed
news_data = news_data.reset_index(drop=True)
```

```python
news_data['label'].value_counts()
```

```python
news_data.shape
```

```python
news_data['content']=news_data['text']+' '+news_data['title']
```

```python
news_data['content']
```

## Stemming

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]',' ',content)
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word)for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)
    return stemmed_content
```

```python
news_data['content']=news_data['content'].apply(stemming)
```

```python
X=news_data['content'].values
Y=news_data['label'].values
```

```python
print(X)
print(Y)
```

## Vectorization

```python
vectorizer=TfidfVectorizer()
```

```python
vectorizer.fit(X)
```

```python
X=vectorizer.transform(X)
```

## Splitting Train Test Data

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)

```python
x_train
```

```python
y_train
```

## Model Training

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

## Making Prediction

```python
prediction=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,prediction)
print("Accuracy = ",accuracy)
```
