## Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

## Importing Dataset

```python
house_price=pd.read_csv('housing.csv')
```

## Data Analysis

```python
house_price=house_price.drop(house_price.columns[0],axis=1)
```

```python
house_price.head()
```

```python
house_price.shape
```

```python
house_price.describe()
```

```python
house_price=house_price.rename(columns={"MEDV":"PRICE"})
```

```python
house_price.head()
```

```python
house_price.isnull().sum()
```

## Identifying Correlation

```python
correlation=house_price.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,
    cmap="Blues",
    annot=True,
    fmt='.1f',
    annot_kws={'size':8},
    cbar=True,
    square="True")
```

## Splitting Training and Testing Data

```python
X=house_price.drop('PRICE',axis=1)
```

```python
Y=house_price['PRICE']
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape,y_test.shape)
```

## Model Training

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

## Predictions

```python
prediction=model.predict(x_test)
```

```python
print(prediction)
```

```python
print("Actual Valuel --> Predicted value")
for i in range (0,20):
    print(y_test.iloc[i],"         --> ",prediction[i])
```

```python
plt.scatter(y_test,prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
plt.plot(np.arange(len(y_test)), sorted(y_test), label='Actual')
plt.plot(np.arange(len(prediction)), sorted(prediction), label='Predicted')
plt.xlabel("Sample Index")
plt.ylabel("Price")
plt.title("Actual vs Predicted Prices (Sorted)")
plt.legend()
plt.show()
```

## Analyzing Accuracy and Error

```python
score_1=metrics.r2_score(y_test,prediction)
score_1
```

```python
score_2=metrics.mean_absolute_error(y_test,prediction)
score_2
```

```python
print("R Squared error = ",score_1)
print("Mean Absolute Error= ",score_1)
```
